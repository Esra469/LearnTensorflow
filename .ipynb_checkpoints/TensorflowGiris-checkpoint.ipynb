## Veriyi test/train olarak ayır

```python
from sklearn.model_selection import train_test_split
```

```python
#y=wx+b
#y->label
y=dataFrame["Fiyat"].values  #Bir numpy dizisine çevirdik (.values diyerek)

# x->feature (özellkler)
x=dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape #verileri ayırdığını görebilrz 670 e 330 olarak ayrım yapılmış
```

# Veriyi normalize etmek 
## Scaling (değerlerin kaydırılıp yeniden ölçeklendirilerek 0 ile 1 arasında bir değere getirilmesiyle oluşan bir ölçekleme tekniğidir)


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train) #bu şekilde veriler 0 ve 1 arasonda bir yere scale edilmiş oldu
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
import tensorflow as tf
```

```python
#tf.keras.layers.Dense()
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
 
model.add(Dense(4,activation="relu")) #tensorflow da o katmanalrın kaç tane olacağını belirliyoruz
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1))

model.compile(optimizer = "rmsprop",loss = "mse") #(mse ->mean squared error)Regressionda çoğunlukla kullanılır
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss=model.history.history["loss"] #lossu dictionary e sonra da diziye döndürüyoruz
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss=model.evaluate(x_train,y_train,verbose=0)
testLoss=model.evaluate(x_test,y_test,verbose=0)
```
